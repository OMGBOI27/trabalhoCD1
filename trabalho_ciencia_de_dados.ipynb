````python
# Merge incidents_master_df with financial_impact_df
merged_df = pd.merge(incidents_master_df, financial_impact_df, on='incident_id', how='left')

# Merge the result with market_impact_df
df = pd.merge(merged_df, market_impact_df, on='incident_id', how='left')

# Display the first 5 rows of the final merged DataFrame and its shape
display(df.head())
print(f"Shape of the final merged DataFrame: {df.shape}")
````

````python
import os
import hashlib
from datetime import datetime

def create_bronze_layer(dataframe, layer_path="/content/datalake/bronze"):
    """
    Creates the Bronze layer by adding ingestion metadata and saving to Parquet.
    """
    # 1. basic Standardization (Column names)
    df_bronze = dataframe.copy()
    df_bronze.columns = [col.lower().replace(' ', '_') for col in df_bronze.columns]

    # 2. Metadata generation
    ingestion_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    source_name = "merged_incidents_data"
    row_count = len(df_bronze)

    # Generate a simple hash of the data content for integrity/versioning
    content_hash = hashlib.md5(df_bronze.to_csv(index=False).encode()).hexdigest()

    df_bronze['ingestion_timestamp'] = ingestion_time
    df_bronze['source_file'] = source_name
    df_bronze['ingestion_row_count'] = row_count
    df_bronze['file_hash'] = content_hash

    # 3. Create directory structure
    os.makedirs(layer_path, exist_ok=True)

    # 4. Save to Parquet
    parquet_file = f"{layer_path}/incidents_raw.parquet"
    df_bronze.to_parquet(parquet_file, index=False)

    print(f"--- Bronze Layer Metadata ---")
    print(f"File: {parquet_file}")
    print(f"Timestamp: {ingestion_time}")
    print(f"Rows: {row_count}")
    print(f"Hash: {content_hash}")

    return df_bronze

# Execute the ingestion to Bronze
df_bronze = create_bronze_layer(df)

# Display the new metadata columns
display(df_bronze[['incident_id', 'ingestion_timestamp', 'source_file', 'file_hash']].head())
````

### 1. Data Lineage: Do CSV à Camada Prata

Este diagrama e descrição representam o ciclo de vida dos dados:

**A. Origens (Raw Sources)**
*   `incidents_master.csv` (Dimensões das empresas e datas)
*   `financial_impact.csv` (Valores de perdas e custos)
*   `market_impact.csv` (Impacto em ações e volatilidade)

**B. Camada Bronze (Ingestão & Padronização)**
*   **Transformação:** Junção das 3 fontes via `incident_id` (Left Join).
*   **Padronização:** Nomes de colunas em *snake_case*.
*   **Metadados:** Adição de `ingestion_timestamp`, `file_hash`, `source_file`.
*   **Persistência:** Salvo em Parquet para preservação de tipos básicos.

**C. Processamento Camada Prata (Cleaning & ML Ready)**
*   **Validações:** Verificação de IDs duplicados e integridade referencial.
*   **Tratamento de Nulos:** Preenchimento de valores financeiros faltantes com 0 ou mediana.
*   **Feature Engineering:**
    *   Criação de `recovery_ratio` (Custo de recuperação / Perda direta).
    *   Cálculo de `days_since_incident`.
    *   One-Hot Encoding de variáveis categóricas críticas.
*   **Remoção:** Colunas de texto livre, URLs e metadados técnicos da Bronze que não servem para o modelo.
*   **Saída:** Dataset final otimizado para treinamento de modelos de impacto financeiro.

### 2. Análise de Qualidade de Dados (Data Quality - Bronze Layer)

Nesta etapa, aplicamos regras automáticas de validação para identificar falhas de integridade nos dados brutos. O objetivo é estabelecer critérios claros de aceitação para o processamento subsequente.

````python
import pandas as pd

def run_bronze_data_quality_report(df):
    results = []

    # 1. Regra: Unicidade do incident_id
    duplicates = df.duplicated(subset=['incident_id']).sum()
    results.append({"Regra": "Unicidade de ID", "Status": "FALHA" if duplicates > 0 else "OK", "Detalhe": f"{duplicates} duplicados encontrados"})

    # 2. Regra: Valores Nulos em Colunas Críticas
    critical_cols = ['incident_id', 'company_name', 'incident_date', 'direct_loss_usd']
    null_counts = df[critical_cols].isnull().sum()
    for col in critical_cols:
        status = "FALHA" if null_counts[col] > 0 else "OK"
        results.append({"Regra": f"Nulos em {col}", "Status": status, "Detalhe": f"{null_counts[col]} valores nulos"})

    # 3. Regra: Padronização de Datas
    try:
        # Tenta converter e identifica erros
        pd.to_datetime(df['incident_date'], errors='raise')
        results.append({"Regra": "Formato de Data", "Status": "OK", "Detalhe": "Todas as datas são válidas"})
    except:
        invalid_dates = pd.to_datetime(df['incident_date'], errors='coerce').isnull().sum()
        results.append({"Regra": "Formato de Data", "Status": "FALHA", "Detalhe": f"{invalid_dates} datas inválidas/fora do padrão"})

    # 4. Regra: Consistência de Categorias (Ex: Industry Primary não deve ser negativo ou vazio)
    if 'industry_primary' in df.columns:
        invalid_cats = df['industry_primary'].astype(str).str.strip().eq('').sum()
        results.append({"Regra": "Categorias de Indústria", "Status": "FALHA" if invalid_cats > 0 else "OK", "Detalhe": f"{invalid_cats} categorias vazias"})

    # Criar DataFrame de resultados para visualização
    df_report = pd.DataFrame(results)

    print("=== RELATÓRIO AUTOMATIZADO DE QUALIDADE (BRONZE) ===")
    display(df_report)

    # Visualização de Nulos Geral
    print("\nPercentual de Nulos por Coluna (Top 10):")
    display((df.isnull().mean() * 100).sort_values(ascending=False).head(10))

    return df_report

# Executar a validação
quality_report = run_bronze_data_quality_report(df_bronze)
````

### 3. Processamento: Camada Prata (ML-Ready)

Esta célula transforma os dados da camada Bronze em um dataset limpo e estruturado, aplicando regras de negócio e preparando os dados para modelos preditivos.

````python
import numpy as np

def build_silver_layer(df_bronze, layer_path="/content/datalake/silver"):
    # 1. Remoção de Duplicidades e Seleção de Colunas (Evitar Data Leakage)
    # Removemos metadados técnicos da Bronze e colunas redundantes ou futuras
    leakage_and_metadata = [
        'ingestion_timestamp', 'file_hash', 'source_file', 'ingestion_row_count',
        'notes', 'notes_x', 'notes_y', 'review_flag', 'data_source_primary', 'data_source_secondary'
    ]
    df_silver = df_bronze.drop_duplicates(subset=['incident_id']).drop(columns=[c for c in leakage_and_metadata if c in df_bronze.columns])

    # 2. Tratamento de Datas
    df_silver['incident_date'] = pd.to_datetime(df_silver['incident_date'])
    df_silver['incident_year'] = df_silver['incident_date'].dt.year
    df_silver['incident_month'] = df_silver['incident_date'].dt.month

    # 3. Padronização de Categorias
    categorical_cols = ['industry_primary', 'attack_vector_primary', 'company_size']
    for col in categorical_cols:
        if col in df_silver.columns:
            df_silver[col] = df_silver[col].str.strip().str.lower().fillna('unknown')

    # 4. Tratamento de Valores Financeiros (Imputação)
    # Se não há valor registrado, assumimos impacto zero para cálculo de métricas agregadas
    fin_cols = ['direct_loss_usd', 'ransom_paid_usd', 'recovery_cost_usd', 'legal_fees_usd', 'regulatory_fine_usd']
    for col in fin_cols:
        if col in df_silver.columns:
            df_silver[col] = df_silver[col].fillna(0)

    # 5. Criação do Label Final (Machine Learning)
    # Definimos como 'Alto Impacto' incidentes com perda direta acima da mediana
    median_loss = df_silver[df_silver['direct_loss_usd'] > 0]['direct_loss_usd'].median()
    df_silver['high_impact_label'] = (df_silver['direct_loss_usd'] > median_loss).astype(int)

    # 6. Feature Engineering Adicional
    df_silver['total_calculated_loss'] = df_silver[fin_cols].sum(axis=1)

    # Salvar Camada Prata
    os.makedirs(layer_path, exist_ok=True)
    output_file = f"{layer_path}/incidents_silver.parquet"
    df_silver.to_parquet(output_file, index=False)

    print(f"--- Camada Prata Concluída ---")
    print(f"Arquivo: {output_file}")
    print(f"Mediana de Perda para Label: ${median_loss:,.2f}")
    print(f"Shape Final: {df_silver.shape}")

    return df_silver

# Execução da transformação
df_silver_final = build_silver_layer(df_bronze)
display(df_silver_final[['incident_id', 'company_name', 'direct_loss_usd', 'high_impact_label']].head())
````

### 6. Relatórios de Qualidade por Camada

Nesta seção, consolidamos a validação de integridade para garantir que cada etapa do pipeline cumpriu seus requisitos de limpeza e transformação.

````python
def generate_comprehensive_report(df_initial, df_bronze, df_silver):
    def get_metrics(df, name):
        return {
            "Camada": name,
            "Total de Linhas": len(df),
            "Total de Colunas": len(df.columns),
            "IDs Duplicados": df.duplicated(subset=['incident_id']).sum() if 'incident_id' in df.columns else 'N/A',
            "Nulos (Direct Loss)": df['direct_loss_usd'].isnull().sum() if 'direct_loss_usd' in df.columns else 'N/A',
            "Metadados Presentes": 'ingestion_timestamp' in df.columns,
            "ML Ready (Label)": 'high_impact_label' in df.columns
        }

    report_data = [
        get_metrics(df_initial, "Inicial (Merge Raw)"),
        get_metrics(df_bronze, "Bronze (Ingestão)"),
        get_metrics(df_silver, "Silver (Limpa/ML)")
    ]

    df_final_report = pd.DataFrame(report_data)

    print("=== RELATÓRIO COMPARATIVO DE QUALIDADE ===")
    display(df_final_report)

    # Verificação de Nulos por Camada
    print("\n--- Evolução de Valores Faltantes (Nulos) em Colunas Críticas ---")
    critical = ['direct_loss_usd', 'recovery_cost_usd', 'incident_date']
    null_evolution = pd.DataFrame({
        'Inicial': df_initial[critical].isnull().sum(),
        'Bronze': df_bronze[critical].isnull().sum(),
        'Silver': df_silver[critical].isnull().sum()
    })
    display(null_evolution)

    return df_final_report

# Gerar o relatório consolidado
final_quality_comparison = generate_comprehensive_report(df, df_bronze, df_silver_final)
````

# 🛡️ Cyber Incident Data Pipeline (Medallion Architecture)

Este notebook implementa um pipeline de dados robusto para processamento de incidentes de cibersegurança, seguindo a arquitetura Medallion (Bronze e Prata) com foco em preparação para Machine Learning.

## 🚀 Instruções de Execução

1.  **Upload de Dados**: Certifique-se de que os arquivos `incidents_master.csv`, `financial_impact.csv` e `market_impact.csv` estão na raiz do diretório `/content/`.
2.  **Setup**: Execute a primeira célula para importar o `pandas`.
3.  **Ingestão & Bronze**: Execute as células de carga e a função `create_bronze_layer`. Isso criará arquivos `.parquet` com metadados de auditoria em `/content/datalake/bronze`.
4.  **Qualidade (DQ)**: Execute o `run_bronze_data_quality_report` para validar a integridade dos dados brutos.
5.  **Refino & Silver**: Execute o `build_silver_layer`. Esta etapa realiza a limpeza, imputação de valores financeiros e criação de labels para modelos preditivos em `/content/datalake/silver`.
6.  **Exploração**: Utilize as células de EDA para visualizar distribuições e correlações financeiras.

---

## 🔍 Checklist Anti-Leakage (Prevenção de Vazamento)

Para garantir que o modelo de ML não utilize informações do 'futuro' ou metadados irrelevantes, o pipeline aplica os seguintes controles na Camada Prata:

| Item | Descrição | Status no Pipeline |
| :--- | :--- | :--- |
| **Metadados Técnicos** | Remoção de `ingestion_timestamp`, `file_hash` e `source_file`. | ✅ Removido |
| **Dados Futuros** | Exclusão de notas de revisão e flags geradas após a análise do incidente. | ✅ Removido |
| **IDs e Chaves** | `incident_id` mantido apenas para referência, não usado como feature. | ✅ Controlado |
| **Imputação Cega** | Preenchimento de nulos financeiros com 0 apenas onde faz sentido técnico. | ✅ Validado |
| **Data Splitting** | Recomendado realizar split temporal (por `incident_year`) antes do treino. | ⚠️ Próximo Passo |
